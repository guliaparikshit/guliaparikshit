Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to

what data visualizations would you use to present your findings?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Google Playstore dataset (replace the file path as needed)
df_playstore = pd.read_csv('google_playstore.csv')

# Preview the dataset to check the columns
print(df_playstore.head())

# Data cleaning: Remove rows with missing values in critical columns
df_playstore = df_playstore.dropna(subset=['Installs', 'Rating', 'Category'])

# Convert 'Installs' to numeric (remove commas and convert to integer)
df_playstore['Installs'] = df_playstore['Installs'].replace({',': ''}, regex=True).astype(int)

# If the dataset contains 'Last Updated' or any time-related column, we can proceed with it
if 'Last Updated' in df_playstore.columns:
    # Convert 'Last Updated' to datetime
    df_playstore['Last Updated'] = pd.to_datetime(df_playstore['Last Updated'], errors='coerce')

    # Extract the month and year from the 'Last Updated' column
    df_playstore['Month'] = df_playstore['Last Updated'].dt.month
    df_playstore['Year'] = df_playstore['Last Updated'].dt.year

    # Group by Month and Year to get total installs by month
    monthly_installs = df_playstore.groupby(['Year', 'Month'])['Installs'].sum().reset_index()

    # Visualization 1: Time series plot for total installs per month
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Month', y='Installs', data=monthly_installs, hue='Year', marker='o')
    plt.title('Total App Installs per Month')
    plt.xlabel('Month')
    plt.ylabel('Total Installs')
    plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.show()

    # Visualization 2: Heatmap of installs by month and year
    monthly_installs_pivot = monthly_installs.pivot('Month', 'Year', 'Installs')
    plt.figure(figsize=(12, 6))
    sns.heatmap(monthly_installs_pivot, annot=True, cmap='YlGnBu', fmt='.0f', cbar=True)
    plt.title('Monthly Install Trends by Year')
    plt.xlabel('Year')
    plt.ylabel('Month')
    plt.show()

# If 'Last Updated' column is not available, perform analysis on other features (e.g., Category)
else:
    print("No 'Last Updated' column found. Analyzing by Category instead.")

    # Group by Category and sum installs
    category_installs = df_playstore.groupby('Category')['Installs'].sum().reset_index()

    # Visualization: Bar chart for installs by category
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Installs', y='Category', data=category_installs.sort_values('Installs', ascending=False).head(10), palette='viridis')
    plt.title('Top 10 App Categories by Total Installs')
    plt.xlabel('Total Installs')
    plt.ylabel('Category')
    plt.show()

```
