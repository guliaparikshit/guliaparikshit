Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python plotting library used to create static, interactive, and animated visualizations.

It's widely used for data visualization in scientific computing, data analysis, and machine learning.

It provides functions to plot various types of graphs.

Five common plots using pyplot from Matplotlib:

Line plot (plt.plot())

Scatter plot (plt.scatter())

Bar plot (plt.bar())

Histogram (plt.hist())

Box plot (plt.boxplot())

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplot with 2 rows and 2 columns
plt.subplot(2, 2, 1)  # First plot
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # Second plot
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # Third plot
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # Fourth plot
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title('Company Profit (Horizontal)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], vert=True, patch_artist=True, labels=["Box 1", "Box 2"])
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```
