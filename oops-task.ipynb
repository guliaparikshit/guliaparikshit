Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle      
        self.max_speed = max_speed                  
        self.average_of_vehicle = average_of_vehicle 

    def display_info(self):
        print(f"Vehicle Name: {self.name_of_vehicle}")
        print(f"Max Speed: {self.max_speed} km/h")
        print(f"Average: {self.average_of_vehicle} km/l")
```

Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle      
        self.max_speed = max_speed                  
        self.average_of_vehicle = average_of_vehicle 

    def display_info(self):
        print(f"Vehicle Name: {self.name_of_vehicle}")
        print(f"Max Speed: {self.max_speed} km/h")
        print(f"Average: {self.average_of_vehicle} km/l")



class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle, capacity):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)
        self.capacity = capacity  

    def seating_capacity(self):
        return f"The {self.name_of_vehicle} has a seating capacity of {self.capacity} people."



my_car = Car("Honda Civic", 220, 18, 5)


my_car.display_info()


print(my_car.seating_capacity())
```

Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance is a feature in Python where a class can inherit from more than one parent class.

```python
# Parent Class 1: Vehicle
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed

    def display_vehicle_info(self):
        print(f"Vehicle Name: {self.name_of_vehicle}")
        print(f"Max Speed: {self.max_speed} km/h")


# Parent Class 2: Engine
class Engine:
    def __init__(self, engine_type):
        self.engine_type = engine_type

    def display_engine_info(self):
        print(f"Engine Type: {self.engine_type}")


# Child Class: Car (Inherits from both Vehicle and Engine)
class Car(Vehicle, Engine):
    def __init__(self, name_of_vehicle, max_speed, engine_type):
        # Initialize Vehicle and Engine classes
        Vehicle.__init__(self, name_of_vehicle, max_speed)
        Engine.__init__(self, engine_type)

    def display_car_info(self):
        self.display_vehicle_info()
        self.display_engine_info()
        print("This is a car with both vehicle and engine characteristics.")


# Creating an object of Car class
car = Car("Tesla Model S", 250, "Electric")

# Display car information
car.display_car_info()
```

Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

Getter: A method used to access the value of a private attribute.
Setter: A method used to set or modify the value of a private attribute.

```python
class Person:
    def __init__(self, name, age):
        self._name = name  # Private attribute 
        self._age = age    # Private attribute 

    # Getter method for 'name'
    @property
    def name(self):
        return self._name

    # Setter method for 'name'
    @name.setter
    def name(self, value):
        if len(value) > 2:  
            self._name = value
        else:
            print("Name must be at least 3 characters long.")

    # Getter method for 'age'
    @property
    def age(self):
        return self._age

    # Setter method for 'age'
    @age.setter
    def age(self, value):
        if value >= 0: 
            self._age = value
        else:
            print("Age cannot be negative.")

# Creating an object of the Person class
person = Person("John", 25)

# Accessing attributes using getter methods
print(person.name)  # Output: John
print(person.age)   # Output: 25

# Modifying attributes using setter methods
person.name = "Alice"  # Valid name change
print(person.name)  # Output: Alice

person.name = "Jo"  
print(person.name)  # Output: Alice

person.age = 30  # Valid age change
print(person.age)  # Output: 30

person.age = -5  
print(person.age)  # Output: 30
```

Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Method overriding in Python occurs when a subclass defines a method with the same name as a method in its parent class. When the subclass method is called, the version in the subclass is executed, overriding the version in the parent class.

```python
# Parent class
class Animal:
    def sound(self):
        print("Some generic animal sound")

# Child class that overrides the sound() method
class Dog(Animal):
    def sound(self):
        print("Bark")

# Child class that overrides the sound() method
class Cat(Animal):
    def sound(self):
        print("Meow")

# Create objects of Animal, Dog, and Cat classes
animal = Animal()
dog = Dog()
cat = Cat()

# Call the sound() method on each object
animal.sound()  
dog.sound()     
cat.sound()     
```
