Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Line plot (sns.lineplot()): Used to display the relationship between two continuous variables, often used for time series data.

Scatter plot (sns.scatterplot()): Used to show the relationship between two continuous variables, helping to identify correlations.

Box plot (sns.boxplot()): Used to visualize the distribution of a continuous variable and detect outliers.

Histogram (sns.histplot()): Used to visualize the distribution of a continuous variable by splitting it into bins.

Heatmap (sns.heatmap()): Used to visualize correlation matrices or data tables, often used to show patterns between variables in large datasets.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Line plot with "timepoint" as x and "signal" as y, for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot of Signal vs Timepoint")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot for age by pclass
plt.figure(figsize=(10, 5))
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Box Plot of Age by Pclass")
plt.show()

# Box plot for fare by pclass
plt.figure(figsize=(10, 5))
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot of Fare by Pclass")
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram of 'price' with hue 'cut'
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)
plt.title("Histogram of Price with Hue by Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Pair plot with hue for species
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to get a matrix form suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, annot=True, cmap="coolwarm")
plt.title("Heatmap of Flight Passengers Over Time")
plt.show()
```
