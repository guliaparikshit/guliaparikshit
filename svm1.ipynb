Q1. What is the mathematical formula for a linear SVM?

In a linear Support Vector Machine (SVM), the model aims to find a straight line (or hyperplane in higher dimensions) that best separates two classes of data. The formula for the SVM's decision boundary is essentially an equation of a line, which tries to place the dividing line as far away as possible from both classes. This ensures the best margin between the classes.

The line or hyperplane is determined by two things:

Weights: The direction of the line.
Bias: The position of the line relative to the origin.


Q2. What is the objective function of a linear SVM?

The goal of a linear SVM is to find the best boundary that separates the two classes. The objective function tries to maximize the margin between the classes while ensuring that the data points are correctly classified. In simple terms, it seeks to draw the line so that it is as far as possible from the closest data points of both classes, called support vectors.

This process also penalizes any mistakes (misclassifications) the model might make, especially when the data is not perfectly separable.

Q3. What is the kernel trick in SVM?

The kernel trick is a technique used to apply SVM to problems where data cannot be separated by a straight line. It allows us to transform the original data into a higher-dimensional space, where it is more likely that a straight line (or hyperplane) can separate the data.

Rather than explicitly mapping the data to a higher dimension, which could be computationally expensive, the kernel trick allows us to compute the necessary transformations efficiently, enabling SVM to handle complex, non-linear decision boundaries.

Q4. What is the role of support vectors in SVM Explain with example

Support vectors are the data points that are closest to the decision boundary (the separating line or hyperplane). These points are critical because they define where the margin should be placed.

Example: Imagine a set of points representing two types of flowers: roses and tulips. If you were to draw a line to separate the two classes, the support vectors would be the flowers that are nearest to that line. These points have the most influence on the position of the line. Removing these support vectors would change the decision boundary, but points further away from the line wouldn't affect the model much.

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

Hyperplane: The line (or flat surface) that separates the two classes. In 2D, this would be a straight line, and in 3D, it would be a plane.

Margin: The distance between the hyperplane and the closest data points from each class (called support vectors). The larger the margin, the better the separation between the classes.

Hard Margin: A strict version where no data points can be on the wrong side of the boundary. This works well only if the data is perfectly separable without any mistakes.

Soft Margin: A more flexible version where some points may fall on the wrong side of the boundary, but the model tries to minimize these errors. This is useful when the data is not perfectly separable.
