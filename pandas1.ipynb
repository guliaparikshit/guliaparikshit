Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Pandas Series: A Series is a one-dimensional array-like object that holds data of a single type. It has an index (default: 0, 1, 2, ...).

Pandas DataFrame: A DataFrame is a two-dimensional table (like a spreadsheet) that consists of multiple columns, each of which can be a Series. It has both row and column indices.

```python
import pandas as pd

series = pd.Series([1, 2, 3, 4])


df = pd.DataFrame({
    'Column1': [1, 2, 3],
    'Column2': ['A', 'B', 'C']
})

print("Series:")
print(series)

print("\nDataFrame:")
print(df)
```

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire', 'David', 'Eve'],
    'Age': [25, 30, 27, None, 22],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

# 1. head() - Show first 3 rows
print("First 3 rows:")
print(df.head(3))

# 2. tail() - Show last 2 rows
print("\nLast 2 rows:")
print(df.tail(2))

# 3. describe() - Summary statistics
print("\nSummary statistics:")
print(df.describe())

# 4. sort_values() - Sort by Age
print("\nData sorted by Age:")
print(df.sort_values(by='Age'))

# 5. drop() - Drop the 'Gender' column
print("\nDataFrame after dropping 'Gender' column:")
print(df.drop('Gender', axis=1))

# 6. groupby() - Group by 'Gender' and get mean age
print("\nAverage Age by Gender:")
print(df.groupby('Gender')['Age'].mean())

# 7. fillna() - Fill missing Age values with mean
print("\nFill missing Age values:")
df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Series: Mutable – You can modify the values, add, or remove data after creation.

DataFrame: Mutable – You can modify the data, add or remove columns/rows, and perform other operations.

Panel: Deprecated – Mutable, but no longer used in modern Pandas. It has been replaced by MultiIndex DataFrames or external libraries like xarray.
